```python
# Identifier les valeurs problÃ©matiques
erreurs = data['serving_size'].apply(lambda x: isinstance(x, str) and len([c for c in x if c == '.']) > 1)
print("ðŸ“Œ Valeurs problÃ©matiques dÃ©tectÃ©es :")
print(data.loc[erreurs, 'serving_size'].head(10))
```

```python
# Identifier les valeurs qui causent problÃ¨me
problemes = data['serving_size'].apply(lambda x: isinstance(x, str) and len(x.strip()) == 0)
print("ðŸ“Œ Valeurs problÃ©matiques (vides ou uniquement espaces) :")
print(data.loc[problemes, 'serving_size'])
```

```python
def clean_serving_size(value):
    if isinstance(value, str):
        # Remplacer les virgules par des points et diviser par espaces
        value = value.replace(',', '.').split()
        
        # Si la liste n'est pas vide, on prend le premier Ã©lÃ©ment
        if len(value) > 0:
            value = value[0]
        else:
            return np.nan
        
        # VÃ©rification du nombre de points dÃ©cimaux
        if value.count('.') > 1:
            # On retire les points en trop
            value = value.replace('.', '', value.count('.') - 1)
        
        # Conversion en float
        try:
            return float(value)
        except ValueError:
            return np.nan
    return value
```

```python
print("ðŸ”„ Nettoyage de la colonne 'serving_size'...")
data['serving_size'] = data['serving_size'].apply(clean_serving_size)

# VÃ©rification aprÃ¨s nettoyage
print("\nðŸ“Œ VÃ©rification des modifications :")
print(data[['serving_size']].head(10))
```

La fonction clean_serving_size a permis de :

Remplacer les virgules par des points dÃ©cimaux,

SÃ©parer les quantitÃ©s des unitÃ©s,

Ã‰liminer les chaÃ®nes non exploitables (NaN pour les cas impossibles Ã  convertir).

```python
import matplotlib.pyplot as plt
import seaborn as sns

# 1ï¸âƒ£ DÃ©finition des bornes logiques pour chaque variable
bornes_logiques = {
    'fiber_100g': (0, 30),                  # Maximum rÃ©aliste pour 100g de produit
    'serving_size': (0, 500),               # 500g semble un bon maximum pour une portion
    'sugars_100g': (0, 100),               # 100g de sucre pour 100g de produit est dÃ©jÃ  Ã©norme
    'fat_100g': (0, 100),                  # Un produit Ã  100% de matiÃ¨res grasses est le maximum logique
    'salt_100g': (0, 10),                  # 10g de sel pour 100g de produit est dÃ©jÃ  excessif
    'sodium_100g': (0, 4),                 # 4g de sodium Ã©quivaut Ã  environ 10g de sel
    'proteins_100g': (0, 80),              # Maximum rÃ©aliste pour 100g de produit
    'energy_100g': (0, 4000)               # 4000 kJ (1000 kcal) pour 100g est dÃ©jÃ  Ã©norme
}

# 2ï¸âƒ£ Fonction de nettoyage
def nettoyer_outliers(df, bornes):
    df_clean = df.copy()
    rapport = []

    print("ðŸš€ Nettoyage des valeurs aberrantes...")

    for colonne, (min_val, max_val) in bornes.items():
        avant = df[colonne].notna().sum()
        df_clean.loc[(df_clean[colonne] < min_val) | (df_clean[colonne] > max_val), colonne] = np.nan
        apres = df_clean[colonne].notna().sum()
        rapport.append((colonne, avant, apres, avant - apres))
    
    # Affichage du rapport
    rapport_df = pd.DataFrame(rapport, columns=['Variable', 'Valeurs initiales', 'Valeurs aprÃ¨s nettoyage', 'Valeurs supprimÃ©es'])
    print("\nðŸ“Š Rapport de nettoyage :")
    print(rapport_df)
    
    return df_clean

# Application de la fonction
X_clean = nettoyer_outliers(X, bornes_logiques)

```

Comparaison visuelle :
Avant nettoyage (en haut) :

On observe de nombreux outliers extrÃªmes, surtout pour les variables fiber_100g, serving_size, sugars_100g et fat_100g.

Certaines valeurs dÃ©passent largement les bornes logiques (ex: 5000g pour fiber_100g, 2500g pour serving_size).

AprÃ¨s nettoyage (en bas) :

Les distributions sont beaucoup plus rÃ©alistes et resserrÃ©es.

Les boxplots montrent un meilleur Ã©talement des donnÃ©es sans les valeurs extrÃªmes.

Lâ€™Ã©chelle est plus cohÃ©rente avec ce quâ€™on peut attendre pour 100g de produit.

Conclusions :
Les valeurs aberrantes ont Ã©tÃ© efficacement supprimÃ©es, rendant les distributions plus rÃ©alistes.
