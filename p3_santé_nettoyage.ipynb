ETAPE 5 : Conversion et nettoyage

```python
# Identifier les valeurs problématiques
erreurs = data['serving_size'].apply(lambda x: isinstance(x, str) and len([c for c in x if c == '.']) > 1)
print("📌 Valeurs problématiques détectées :")
print(data.loc[erreurs, 'serving_size'].head(10))
```

```python
# Identifier les valeurs qui causent problème
problemes = data['serving_size'].apply(lambda x: isinstance(x, str) and len(x.strip()) == 0)
print("📌 Valeurs problématiques (vides ou uniquement espaces) :")
print(data.loc[problemes, 'serving_size'])
```

```python
def clean_serving_size(value):
    if isinstance(value, str):
        # Remplacer les virgules par des points et diviser par espaces
        value = value.replace(',', '.').split()
        
        # Si la liste n'est pas vide, on prend le premier élément
        if len(value) > 0:
            value = value[0]
        else:
            return np.nan
        
        # Vérification du nombre de points décimaux
        if value.count('.') > 1:
            # On retire les points en trop
            value = value.replace('.', '', value.count('.') - 1)
        
        # Conversion en float
        try:
            return float(value)
        except ValueError:
            return np.nan
    return value
```

```python
print("🔄 Nettoyage de la colonne 'serving_size'...")
data['serving_size'] = data['serving_size'].apply(clean_serving_size)

# Vérification après nettoyage
print("\n📌 Vérification des modifications :")
print(data[['serving_size']].head(10))
```

La fonction clean_serving_size a permis de :

Remplacer les virgules par des points décimaux,

Séparer les quantités des unités,

Éliminer les chaînes non exploitables (NaN pour les cas impossibles à convertir).

______________

Les valeurs comme 28 g, 35 g, 52 g ont été correctement converties en nombres (float).

Les valeurs problématiques que nous avions détectées (comme "1.4 g (0.25 tsp)") ont été converties en NaN, ce qui est une bonne chose car elles nécessitent un traitement plus poussé.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Liste des variables quantitatives
variables_quantitatives = X.select_dtypes(include=['float64', 'int64']).columns

# Dimensions de l'affichage
plt.figure(figsize=(15, 12))

# Création des boxplots pour chaque variable quantitative
for i, col in enumerate(variables_quantitatives, 1):
    plt.subplot(5, 3, i)
    sns.boxplot(x=X[col], color='lightblue')
    plt.title(f"Distribution de '{col}'")
    plt.tight_layout()

plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# 1️⃣ Définition des bornes logiques pour chaque variable
bornes_logiques = {
    'fiber_100g': (0, 30),                  # Maximum réaliste pour 100g de produit
    'serving_size': (0, 500),               # 500g semble un bon maximum pour une portion
    'sugars_100g': (0, 100),               # 100g de sucre pour 100g de produit est déjà énorme
    'fat_100g': (0, 100),                  # Un produit à 100% de matières grasses est le maximum logique
    'salt_100g': (0, 10),                  # 10g de sel pour 100g de produit est déjà excessif
    'sodium_100g': (0, 4),                 # 4g de sodium équivaut à environ 10g de sel
    'proteins_100g': (0, 80),              # Maximum réaliste pour 100g de produit
    'energy_100g': (0, 4000)               # 4000 kJ (1000 kcal) pour 100g est déjà énorme
}

# 2️⃣ Fonction de nettoyage
def nettoyer_outliers(df, bornes):
    df_clean = df.copy()
    rapport = []

    print("🚀 Nettoyage des valeurs aberrantes...")

    for colonne, (min_val, max_val) in bornes.items():
        avant = df[colonne].notna().sum()
        df_clean.loc[(df_clean[colonne] < min_val) | (df_clean[colonne] > max_val), colonne] = np.nan
        apres = df_clean[colonne].notna().sum()
        rapport.append((colonne, avant, apres, avant - apres))
    
    # Affichage du rapport
    rapport_df = pd.DataFrame(rapport, columns=['Variable', 'Valeurs initiales', 'Valeurs après nettoyage', 'Valeurs supprimées'])
    print("\n📊 Rapport de nettoyage :")
    print(rapport_df)
    
    return df_clean

# Application de la fonction
X_clean = nettoyer_outliers(X, bornes_logiques)

```

Comparaison visuelle :
Avant nettoyage (en haut) :

On observe de nombreux outliers extrêmes, surtout pour les variables fiber_100g, serving_size, sugars_100g et fat_100g.

Certaines valeurs dépassent largement les bornes logiques (ex: 5000g pour fiber_100g, 2500g pour serving_size).

Après nettoyage (en bas) :

Les distributions sont beaucoup plus réalistes et resserrées.

Les boxplots montrent un meilleur étalement des données sans les valeurs extrêmes.

L’échelle est plus cohérente avec ce qu’on peut attendre pour 100g de produit.

Conclusions :
Les valeurs aberrantes ont été efficacement supprimées, rendant les distributions plus réalistes.

```python
import pandas as pd

# 🚀 Imputation des valeurs manquantes par type de variable
print("🔄 Imputation des valeurs manquantes...")

# 1️⃣ Variables numériques : imputation par la médiane
variables_numeriques = [
    'fiber_100g', 'carbohydrates_100g', 'fat_100g', 
    'sugars_100g', 'saturated-fat_100g', 'salt_100g', 
    'sodium_100g', 'proteins_100g', 'energy_100g'
]
for var in variables_numeriques:
    median_value = X_clean[var].median()
    X_clean[var].fillna(median_value, inplace=True)
    print(f"✅ Imputation de la médiane pour {var} : {median_value}")

# 2️⃣ Variables catégorielles : imputation par le mode
variables_categorielles = [
    'serving_size', 'additives', 'ingredients_text', 
    'brands_tags', 'brands', 'product_name', 
    'countries_fr', 'countries_tags', 'countries'
]
for var in variables_categorielles:
    mode_value = X_clean[var].mode()[0]
    X_clean[var].fillna(mode_value, inplace=True)
    print(f"✅ Imputation du mode pour {var} : {mode_value}")

# 3️⃣ Variables booléennes : imputation par 0
variables_booleennes = [
    'ingredients_from_palm_oil_n', 'ingredients_that_may_be_from_palm_oil_n', 
    'additives_n'
]
for var in variables_booleennes:
    X_clean[var].fillna(0, inplace=True)
    print(f"✅ Imputation par 0 pour {var}")

# 🔍 Vérification finale des valeurs manquantes
valeurs_manquantes_apres = X_clean.isnull().sum().sum()
print(f"📌 Valeurs manquantes restantes après imputation : {valeurs_manquantes_apres}")

```
